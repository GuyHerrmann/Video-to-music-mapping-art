```python
contours, hierarchy = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
print('#contours: {}'.format(len(contours)))
contours = np.array(sorted(contours, key=len, reverse=True))

```

```python
size_list = np.array([len(i) for i in contours])

assert size_list.all() == np.sort(size_list).all()

threshold_value = 60
above_val = np.count_nonzero(size_list> threshold_value)

sorted_contours = contours[:above_val]
print([len(i) for i in sorted_contours])
```

```python
def map_coords(coord, xdim, ydim, xint, yint):
    x = int(((coord[0]/xdim)*xint)+0.5)
    y = int(((coord[1]/ydim)*yint)+0.5)
    return (x,y)
```

```python
d_points = dict()
for i, point in enumerate(sorted_contours):
    _area = len(point)
    _tmp, _rad = cv2.minEnclosingCircle(point)
    _coord = map_coords(_tmp, x_dim, y_dim, x_inter, y_inter)
    d_points[i] = (_area, _rad, _coord)

```
